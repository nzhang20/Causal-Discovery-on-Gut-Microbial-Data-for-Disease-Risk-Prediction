# Get all the studies for which we have feature table data

```python
otu_table = biom.load_table("../data/sam/genus-table-exported/feature-table.biom")
otu_table
```

```python
metadata = pd.read_csv("../birdman/metadata_2024Feb02.tsv", sep="\t", index_col=0)
metadata
```

```python
metadata_w_otu = metadata.loc[otu_table.to_dataframe().columns.intersection(metadata.index)].drop(
    columns=['Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Diet'])
metadata_w_otu
```

# PN.902.a `metadata_w_otu` Disease == Healthy

```python
healthy_samples = metadata_w_otu[metadata_w_otu['Disease'] == 'Healthy']
pd.concat([healthy_samples['Study'].value_counts(), healthy_samples.groupby('Study').first()], axis=1)
```

# PN.902.b `metadata_w_otu` Disease == T2D, Obesity/T2D, Type II Diabetes, Type 2 diabetes, Type_II_Diabetes

```python
diabetes_cat = ['T2D', 'Obesity/T2D', 'Type II Diabetes', 'Type 2 diabetes', 'Type_II_Diabetes']
t2d_samples = metadata_w_otu[metadata_w_otu['Disease'].isin(diabetes_cat)]
pd.concat([t2d_samples['Study'].value_counts(), t2d_samples.groupby('Study').first()], axis=1)
```
